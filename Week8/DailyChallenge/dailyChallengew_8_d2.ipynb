#‚úÖ √âtape 4 : Charger le mod√®le avec llama-cpp-python et g√©n√©rer du texte
L'Objectif est de :

Utiliser le mod√®le .gguf t√©l√©charg√© pour :

r√©pondre √† une question en langage naturel,

g√©n√©rer un script Python automatiquement.



```python
#
# Importer la classe principale pour interagir avec le mod√®le
from llama_cpp import Llama

# Initialiser l'objet Llama avec le bon chemin de mod√®le
llm = Llama(
    model_path="gemma-3b.Q8_0.gguf",  # Fichier GGUF t√©l√©charg√© pr√©c√©demment
    n_gpu_layers=32,                  # Nombre de couches sur GPU (tu peux ajuster selon ta VRAM dispo)
    n_threads=8                       # Nombre de threads CPU √† utiliser
)
```

#√âtape 4.2 : G√©n√©rer une r√©ponse en langage naturel

```python
prompt = "Explique comment le syst√®me solaire s'est form√©."

# G√©n√©ration du texte
output = llm(prompt=prompt, max_tokens=200)  # 200 tokens ‚âà 150 mots
print(output["choices"][0]["text"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def trace_line_chart(csv_file, title="Graph", x_label="X", y_label="Y"):
    """
    Lit un fichier CSV et trace un graphique en ligne (line chart) avec matplotlib.

    Args:
        csv_file (str): Le chemin vers le fichier CSV.
        title (str): Le titre du graphique.
        x_label (str): Le label pour l'axe des x.
        y_label (str): Le label pour l'axe des y.
    """
    try:
        df = pd.read_csv(csv_file)
    except FileNotFoundError:
        print(f"Erreur : Le fichier '{csv_file}' n'a pas √©t√© trouv√©.")
        return
    except Exception as e:
        print(f"Erreur lors de la lecture du fichier CSV : {e}")
        return

    try:
        df.plot(x=df.columns[0], y=df.columns[1])
        plt.title(title)
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.show()
    except Exception as e:
        print(f"Erreur lors de la cr√©ation du graphique : {e}")
        return
```

```python
# Cr√©ation du fichier data :un petit CSV de test (2 colonnes : Jour, Valeur)
import pandas as pd

df = pd.DataFrame({
    "Jour": [1, 2, 3, 4, 5],
    "Valeur": [10, 12, 9, 14, 11]
})

df.to_csv("data.csv", index=False)
```

```python
# Exemple d‚Äôappel (√† adapter selon le nom de ton fichier)
trace_line_chart("data.csv", title="Courbe de test", x_label="Jour", y_label="Valeur")
```
