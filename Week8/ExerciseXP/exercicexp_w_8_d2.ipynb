## Exercice 1 – Étape 1 : Gather Definitions
📌 Objectif de l’étape : Rechercher et noter les définitions des trois niveaux d’ouverture d’un modèle LLM.

* Fully Open : Le code source, l’architecture du modèle, et les poids

préentraînés sont disponibles publiquement et sans restriction majeure.

* Weights Released : Les poids du modèle sont accessibles (souvent pour l’affinage), mais le code source ou les données d'entraînement ne sont pas forcément partagés.

Architecture Only : Seule la structure (layers, dimensions, attention mechanism, etc.) du modèle est publiée, sans les poids préentraînés.

##Exercice 1 – Étape 2 : Identify Key Characteristics

1. Fully Open
✅ Ce qui est ouvert : Code source, poids préentraînés, architecture complète.

⚠️ Ce que tu peux faire : Tout – inspection, modification, réentraînement, déploiement commercial.

2. Weights Released
✅ Ce qui est ouvert : Poids préentraînés uniquement.

⚠️ Ce que tu peux faire : Fine-tuning possible, mais accès limité à la logique interne du code ou aux données d’origine.

3. Architecture Only
✅ Ce qui est ouvert : Description de l’architecture du modèle (schéma, dimensions, couches…).

⚠️ Ce que tu peux faire : Impossible d’utiliser ou d’affiner sans réentraîner les poids à partir de zéro.



##Exercice 1 – Étape 3 : Compare Them Side by Side

| Niveau d’ouverture | Qu’est-ce qui est ouvert ?           | Impact sur l'entraînement / modification                         |
| ------------------ | ------------------------------------ | ---------------------------------------------------------------- |
| Fully Open         | Code + poids + architecture          | Tout est possible : inspection, fine-tuning, déploiement         |
| Weights Released   | Poids uniquement                     | Fine-tuning possible, mais pas de modification de l’architecture |
| Architecture Only  | Structure du modèle (sans les poids) | Réentraînement complet requis ; pas de fine-tuning possible      |


#Exercice 1 – Étape 4 : Rédiger un paragraphe comparatif

Objectif : Résumer les trois niveaux d’ouverture en 3 à 5 phrases, en insistant sur leurs avantages ou limites.

Les niveaux d’ouverture des LLM varient selon les composants accessibles. Un modèle Fully Open met à disposition son code, ses poids préentraînés et son architecture, ce qui permet une transparence totale, une modification complète et un réentraînement sans restriction. Un modèle avec uniquement les Weights Released offre ses poids pour un fine-tuning possible, mais le code source ou l’architecture détaillée peuvent manquer, limitant la personnalisation. Enfin, un modèle avec Architecture Only ne fournit que le schéma du modèle, sans les poids, ce qui empêche tout usage direct ou fine-tuning, à moins de réentraîner depuis zéro. Le degré d’ouverture choisi influence donc fortement ce qu’on peut faire avec le modèle dans un contexte pratique.



##Exercice 1 – Étape 5 : Répondre au cas d’usage en santé

Le niveau Fully Open est essentiel pour ce cas, car il permet un accès complet au modèle pour le fine-tuning sur des données cliniques. Ce niveau garantit une transparence et un contrôle total, ce qui est crucial pour respecter les exigences de confidentialité et de conformité dans le domaine médical.

#Étape 2 : Liste des Pour / Contre (5 bullet points)

Voici une comparaison équilibrée que tu peux reprendre ou adapter :

🖥️ Local Deployment
✔️ Faible latence, pas de cold start (le modèle est toujours dispo)

❌ Consomme beaucoup de RAM et CPU/GPU local

✔️ Contrôle total sur les données (pas d’envoi à des serveurs tiers)

❌ Configuration initiale complexe (installations, compilation, compatibilité)

❌ Limité par ton matériel (taille de modèle, vitesse d'inférence)

☁️ Cloud Deployment
✔️ Facile à démarrer (Google Colab, RunPod, Hugging Face Spaces)

✔️ Accès rapide à des GPUs puissants sans investissement

❌ Latence potentielle (temps de démarrage, files d'attente)

❌ Risques de confidentialité (données transitant par le cloud)

❌ Coût à long terme (abonnements, crédits GPU, limites horaires)

```python
!pip install transformers accelerate

from transformers import AutoModelForCausalLM, AutoTokenizer
import torch, time

model_name = "meta-llama/Llama-2-7b-chat-hf"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", torch_dtype=torch.float16)

inputs = tokenizer("Hello, world!", return_tensors="pt").to(model.device)
start = time.time()
outputs = model.generate(**inputs, max_new_tokens=50)
print(tokenizer.decode(outputs[0]))
print("Elapsed:", time.time() - start, "seconds")
```
